```python
%matplotlib notebook

#Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
from scipy.stats import linregress
import os
import csv
```

```python
# define variables, lists
# years of data for analysis
year_list = ['2015', '2016', '2017', '2018', '2019', '2022']
test_list = ['2015-2019', '2022'] #For class presentation
# project happiness survey categories
categories = ['Happiness Score', 'Family', 'Health', 'Freedom', 'Perceptions of Corruption', 'Generosity']

## assumption - complete happiness is an overall score of 10 with 1/5 or 2 points being the ideal for each category
```

```python
# Load combines csv file containing all the data from 2015, 2016, 2017, 2018, 2019 and 2022
combo_data_df = pd.read_csv('Output_data/Happiness_df.csv')

combo_data_df
```

```python
#Clear out null values. 
combo_data_df.isnull().values.any()
```

```python
# Add Freedom average combined results for 2015-2019, add column

combo_data_df['Freedom (2015-2019)']=((combo_data_df['Freedom (2015)']+
                                       combo_data_df['Freedom (2016)']+
                                       combo_data_df['Freedom (2017)']+
                                       combo_data_df['Freedom (2018)']+
                                       combo_data_df['Freedom (2019)']
                                                    )/5)
#Display happiness scores_df.
combo_data_df
```

```python
# Get average Overall Happiness scores for 2015-2019, add column.

combo_data_df['Overall Happiness (2015-2019)']=((combo_data_df['Happiness Score (2015)']+
                                       combo_data_df['Happiness Score (2016)']+
                                       combo_data_df['Happiness Score (2017)']+
                                       combo_data_df['Happiness Score (2018)']+
                                       combo_data_df['Happiness Score (2019)']
                                                    )/5)
#Display happiness scores_df.
combo_data_df
```

```python
# Create Scatter Plots of Freedom vs Overall Happiness

loop_category = categories[3] # Freedom Happiness Score
top_category = categories[0]  # Overall Happiness Score

i = 0
for year in test_list:
    test_year = test_list[i]
    ## create the column for testing
    col = f'{loop_category} ({test_year})'
    ## create the column for testing
    top_col = f'{top_category} ({test_year})'    
    ## create the DataFrame for testing

# sort by the category value and create DataFrame for plotting
#combo_data_df_sorted = combo_data_df.sort_values(f'{top_col}', ascending=False)    

# for Overall Freedom Score (x-values) and Category Scores (y_values) for the Years selected
# values for plotting
x_values = combo_data_df.loc[f'{top_col}']
y_values = combo_data_df.loc[f'{col}']

# statistics
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values =x_values * slope + intercept

# set up the y=mx+b detail to overlay the scatter plot
slope_intercept = 'y = ' + str(round(slope,2)) + 'x +' + str(round(intercept,2))  ## y = mx + b

# plot details, add annotation, add chart features
plt.figure()
plt.scatter(x_values, y_values, s=150, marker='o', edgecolors='Black', c='Green')
plt.plot(x_values, regress_values, 'r-') ##'r-' is a red line
    
plt.annotate(slope_intercept, (10,10), fontsize=20, color="red")
    
#Set the graph title, labels, grid, save findings.
plt.title(f'{loop_category} vs Overall Happiness Scores for {test_year}', fontsize=16, weight='bold')           
plt.xlabel(f'{top_col} score', fontsize=14, weight='bold')
plt.ylabel(f'{col} score', fontsize=14, weight='bold')
plt.grid(color='gray', linestyle='-', linewidth=0.2)
    
#Print values on graph.
#print(f'For the Happiness Survey Year(s) {"test_year}')
#print(f' The r-value isL {round(rvalue**2,3)}and the pvalue is: {round9pvalue**2,3}')
#print(f'The slope intercept is: {slope_intercept}')

#Save graphs.
#plt.savefig("output data/Fig12.png")
i += 1
plt.show()
```
